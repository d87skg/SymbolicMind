```
# 安装 SymbolicMind
!pip install -q symbolimind numpy scipy scikit-learn torch
import sys
from symbolimind.unified import SymbolicMind
sm = SymbolicMind(load_extensions=True)
print('SymbolicMind 已就绪')
```

```
# 上传你的 CSV 数据并运行
from google.colab import files
uploaded = files.upload()
import io
for fn in uploaded.keys():
    csv_data = uploaded[fn].decode('utf-8')
    result = sm.discover(data_csv=csv_data, mode='time_series')
    print(f'文件: {fn}')
    print(f'方程: {result["equation"]}')
    print(f'R²: {result["r2"]:.4f}')
    print(f'P0: {result["p0_verdict"]}')
```
